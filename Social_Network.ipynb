```python
class SocialNetwork:
    def __init__(self):
        self.users = {}

    def register(self, username, password):
        if username in self.users:
            print("‚ùå Username already exists")
        else:
            self.users[username] = {
                "password": password,
                "friends": [],
                "posts": []
            }
            print("‚úÖ Registration successful")

    def login(self, username, password):
        if username in self.users and self.users[username]["password"] == password:
            print("‚úÖ Login successful")
            return username
        else:
            print("‚ùå Invalid credentials")
            return None

    def add_friend(self, user, friend):
        if friend in self.users:
            if friend not in self.users[user]["friends"]:
                self.users[user]["friends"].append(friend)
                print("‚úÖ Friend added")
            else:
                print("‚ö†Ô∏è Already friends")
        else:
            print("‚ùå User not found")

    def post_status(self, user, message):
        self.users[user]["posts"].append(message)
        print("‚úÖ Status posted")

    def view_feed(self, user):
        print("\nüì∞ News Feed")
        for friend in self.users[user]["friends"]:
            posts = self.users[friend]["posts"]
            for post in posts:
                print(f"{friend}: {post}")

    def view_friends(self, user):
        print("\nüë• Friends List:")
        for friend in self.users[user]["friends"]:
            print(friend)


# ================= MAIN PROGRAM =================

sn = SocialNetwork()
current_user = None

while True:
    print("\n--- Social Network Menu ---")
    print("1. Register")
    print("2. Login")
    print("3. Add Friend")
    print("4. Post Status")
    print("5. View Feed")
    print("6. View Friends")
    print("7. Logout")
    print("8. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        u = input("Username: ")
        p = input("Password: ")
        sn.register(u, p)

    elif choice == "2":
        u = input("Username: ")
        p = input("Password: ")
        current_user = sn.login(u, p)

    elif choice == "3" and current_user:
        f = input("Friend username: ")
        sn.add_friend(current_user, f)

    elif choice == "4" and current_user:
        msg = input("Enter status: ")
        sn.post_status(current_user, msg)

    elif choice == "5" and current_user:
        sn.view_feed(current_user)

    elif choice == "6" and current_user:
        sn.view_friends(current_user)

    elif choice == "7":
        current_user = None
        print("üëã Logged out")

    elif choice == "8":
        print("üëã Goodbye!")
        break

    else:
        print("‚ö†Ô∏è Please login first or choose a valid option")
```
